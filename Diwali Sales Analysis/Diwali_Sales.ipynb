# Diwali Sales Data 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
file_path= r"D:\PG_DS\Python\DataSets\Diwali_sales\Diwali Sales Data.csv"
data=pd.read_csv(file_path, encoding= 'unicode_escape')
```

```python
data.head()
```

```python
diwali=data.copy()
```

```python
diwali.head()
```

```python
diwali.info()
```

```python
diwali.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
diwali.info()
```

```python
diwali.dropna(inplace=True)
```

```python
diwali.info()
```

```python
diwali['Amount']=diwali['Amount'].astype('int')
```

```python
diwali.columns
```

```python
diwali.describe()
```

# EDA

Gender

```python
sns.countplot(x='Gender',data=diwali)
```

```python
gen_sales=diwali.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)

sns.barplot(x="Gender",y="Amount",data=gen_sales)
```

Females have more purchasing power and most of the buyers are female

Age

```python
desired_order=['0-17','18-25','26-35','36-45','46-50','51-55','55+']

ax=sns.countplot(x='Age Group',hue='Gender',data=diwali,order=desired_order)

for bars in ax.containers:
    ax.bar_label(bars)
```

Most of the buyers are Female from Age Group= 26-35 

```python
age_sales=diwali.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)

ax=sns.barplot(x="Age Group",y="Amount",data=age_sales,palette='rainbow')
```

Most of the buyers are from age group 26-35 and are females

# State

```python
state_sales=diwali.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.barplot(x='State',y='Orders',data=state_sales,palette='rainbow')

plt.tight_layout()

plt.show()
```

```python
state_sales_amt=diwali.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.barplot(x='State',y='Amount',data=state_sales_amt,palette='rainbow')

plt.tight_layout()

plt.show()
```

Most of the orders and sales are from UP,MH and KA

# Occupation

```python
ax = sns.countplot(data = diwali, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.tight_layout()

plt.show()
```

```python
sales_occup=diwali.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Occupation',y='Amount',data=sales_occup,palette='rainbow')

plt.tight_layout()

plt.show()
```

Most buyers are working in IT,Healthcare and Aviation

# Marital Status 

```python
ax = sns.countplot(data = diwali, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
marital_sales = diwali.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)


sns.barplot(data = marital_sales, x = 'Marital_Status',y= 'Amount', hue='Gender',palette='rainbow')
```

Most of the buyers are Female- Unamrried and followed by Female Married

# Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = diwali, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.tight_layout()

plt.show()
```

```python
sales_products = diwali.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_products, x = 'Product_Category',y= 'Amount',palette='rainbow')
```

Most purchased products are from Food,Clothing ,Electronics,Footwear respectively
